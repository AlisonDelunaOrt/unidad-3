Plotting sin linea
Para trazar solo los marcadores,puede usar notacion de cadena de acceso directo 'o',que significa "anillos"
Ejemplo:Dibuja dos puntos en el diagrama,uno en la posicion(1,3) y otro en la posicion (8,10)

```python
import numpy as np
import matplotlib.pyplot as plt
xpuntos=np.array([1,8])
ypuntos=np.array([3,10])
plt.plot(xpuntos,ypuntos,'o')
plt.show()
```

Diagrama de dispercion (scatter)

Se emplea para mostrar la relacion entre dos variables numericas,detectando correlaciones o agrupamientros,
Ejemplo:


```python
import matplotlib.pyplot as plt

altura = [150, 160, 170, 180, 190, 200]
peso = [50, 60, 65, 80, 85, 95]

plt.scatter(altura, peso,color = "red", marker = "x")
plt.title("Relación entre Altura y Peso")
plt.xlabel("Altura(cms)")
plt.ylabel("Peso(kg)")
plt.show()
```

Graficos de barras (bar y bahr)
Sirven para comparar categorias o variables tanto en orientacion vertical como horizontal.
Ejemplo de barras verticales:

```python
import matplotlib.pyplot as plt

categorias = ["A", "B", "C", "D"]
valores = [4, 7, 1, 8]

plt.bar(categorias, valores,color = "orange")
plt.title("Gráfico de Barras Verticales")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```
