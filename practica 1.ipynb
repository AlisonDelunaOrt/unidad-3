Matplotlib pyplot
Pyplot
La mayoria de las utlidades de Matplotlib se encuentran bajo pyplot submodulo,y normalmente se importan bajo el plt alias:
import matplotlib.pyplot as plt
Ahora el paquete pyplot puede denominarse plt.
Ejemplo:Dibuja una linea en un diagrama desde la posicion (0,0) hasta la posicion (6,250)

```python
import matplotlib.pyplot as plt
import numpy as np
xpuntos = np.array([0, 6])
ypuntos = np.array([0, 250])
plt.plot(xpuntos, ypuntos)
plt.show()
```

Matplotlib plotting
Trazar puntos x - y
La funcion plot() se utiliza para dibujar puntos(marcadores) en un diagrama por defecto la funcion plot() dibuja una linea de un punto al otro.
Esta funcion toma parametros para especificar los puntos en el diagrama.
El primer parametro es una matriz que contiene los puntos del eje x.En el segundo parametro es una matriz que contiene los puntos del eje Y.
Si necesitamos trazar una linea de (1,3) a (8,10),tenemos que pasar estos datos en 2 matrices[1,8] y [3,10] a la funcion del trazado plot().
Ejemplo:Dibuja una linea en un diagrama desde la p√≥sicion (1,3)hasta la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np
xpuntos = np.array([1, 8])
ypuntos = np.array([8, 10])
plt.plot(xpuntos, ypuntos)
plt.show()
```
