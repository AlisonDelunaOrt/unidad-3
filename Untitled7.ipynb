Conceptos basicos de la libreria Matplotlib
-------------------------------
Matplotlib es una libreria de transados de graficas de bajo nivel de visualizacion .Matplotlib es de codigo abierto y podemos utilizarlo libremente.Matplotlib esta escrito principalmente en python con algunos segmentos escritos en C.
-------------------------
Instalacion de matplotlib.Si tienes python y pip ya instalado en tu sistema,instala matplotlib usando el siguiente comando:
----------------
pip install matplotlib
     desde la linea de comando.
---------------------------
Import matplotlib.Una vez instalado matplotlib,importalo en tus aplicaciones agregando el modulo import.
-------------
Ejemplo:matplotlib pyplot

------------------
Pyplot:La mayoria de las herramientas de matplotlib se encuentran dentro del submodulo pyplot y normalmente se importan bajo el alias plt:
-------------
import matplotlib.pyplot as plt
Ahora el paquete puede denominarse plt
